## **Excel file Format**

In XLSX data is organized under the cells and columns in a sheet.

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00390/stock portfolio performance data set.xlsx"
url = url.replace(" ", "%20")
portfolio_performance = pd.read_excel(url,header=0)
portfolio_performance.head(10)
```

```python
portfolio_performance.columns = ["ID", "X1", "X2", "X3", "X4", "X5", "X6", "Y1",
                                 "Y2", "Y3", "Y4", "Y5", "Y6", "NY1", "NY2", "NY3",
                                 "NY4", "NY5", "NY6"]

portfolio_performance.head(10)
```
